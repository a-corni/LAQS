```julia
#Define space :

# Description of an experiment in a cavity
# The experiment here is in free space, uncomment to place yourself in the conditions of this experiment

Lx = 20*10^(-2); #length of the cavity along x-axis - in m
Ly = 20*10^(-2); #length of the cavity along y-axis - in m
Lz = 10^(-1); #lenth of the cavity along z-axis - in m
#The origin is placed at the center of the cavity
Bx = Lx/2; #extreme point along x axis
By = Ly/2; #extreme point along y axis
Bz = Lz/2; #extreme point along z axis

#General Physical constants:

mu0 = 4*pi*10^(-7) #T.m/A
k_B = 1.3806504*10^(-23) #J/K
e = 1.602176487*10^(-19) #C
a0 = 0.52917720859*10^(-10)#m #Bohr Radius
c = 299792458 #m/s
h = 6.62607004*10^(-34) #m².kg/s

#Initial Properties of Cesium gaz:

m = 2.2*10^(-25) #kg
T = 1 #°K
N = 10 #Number of Cesium atoms

#Model Cesium atoms as 2-level atoms:
#Transition Cs |6S(1/2) F = 4> -> Cs |6P(3/2) F = 5>

wa = 2*pi*351.72571850*10^(12) #rad.s-1 #transition frequency
γ = 2*pi*5.18*10^6 #Hz #decay rate
tau = 30.405*10^(-9) #s #lifetime
vr = 3.5225 * 10^(-3) #m.s-1 #recoil velocity
wr = 2*pi*2.0663*10^3 #Hz #recoil "energy"
Tr = 198.34 * 10^(-9) #K #recoil Temperature
Td = 125.61*10^(-6) #K #doppler Temperature
d = 2.5886*e*a0
Is = 11.049 #W/m^2 #Staturation Intensity for a σ+- light for a Cs |6S(1/2) F = 4, mF = +-4> -> Cs |6P(3/2) F = 5, mF = +-5>

#Beam properties:

Dl = 11*10^(-2) #m #diameter of the beam at the origin
δ = -0.5*γ #s-1 #detuning frequency
wl = wa + δ #rad.s-1 #laser frequency
λ = 2*pi*c/wl #m #laser wavelength
k = wl/c #rad.m-1 #laser wave number
I0 = 0.1*Is #intensity at 
w0 = Dl/2 #cm 
zR = pi*w0^2/λ
ex = [1, 0, 0]
ey = [0, 1, 0]
ez = [0, 0, 1]

function Ix_minus(x, y,z) #Intensity beam A1
  #Beam A1 : sigma - polarized transverse gaussian light propagating in the direction -ex
  #maximum intensity reached at the origin  
    return I0/abs(1+(x/zR)^2)*exp(-2*(y^2+z^2)/w0^2/(1+(x/zR)^2))    
end

function Ix_plus(x, y,z) #Intensity beam A2
  #Beam A2 : sigma + polarized transverse gaussian light propagating in the direction +ex  
  #maximum intensity reached at the origin
    return I0/abs(1+(x/zR)^2)*exp(-2*(y^2+z^2)/w0^2/(1+(x/zR)^2))    
end

function Iy_plus(x, y,z) #Intensity beam B2
  #Beam B2 : sigma + polarized transverse gaussian light propagating in the direction +ey 
  #maximum intensity reached at the origin
    return I0/abs(1+(y/zR)^2)*exp(-2*(z^2+x^2)/w0^2/(1+(y/zR)^2))    
end

function Iy_minus(x, y,z) #Intensity beam B1
  #Beam B1 : sigma - polarized transverse gaussian light propagating in the direction -ey  
  #maximum intensity reached at the origin
    return I0/abs(1+(y/zR)^2)*exp(-2*(z^2+x^2)/w0^2/(1+(y/zR)^2))    
end

function Iz_plus(x, y,z) #Intensity beam C2
  #Beam C2 : sigma + polarized gaussian light propagating in the direction +ez  
  #maximum intensity reached at the origin
    return I0/abs(1+(z/zR)^2)*exp(-2*(x^2+y^2)/w0^2/(1+(z/zR)^2))    
end

function Iz_minus(x, y,z) #Intensity beam C1
  #Beam C1 : Transverse gaussian sigma - polarized light propagating in the direction -ez  
  #maximum intensity reached at the origin
    return I0/abs(1+(z/zR)^2)*exp(-2*(x^2+y^2)/w0^2/(1+(z/zR)^2))    
end

function Ω(x, y, z) #Rabi frequency
    return sqrt(I(x, y, z)/Is/2)*γ
end

function sA2(x, y, z) #saturation parameter Beam A2
    return Ix_plus(x, y, z)/Is
end

function sA1(x, y, z) #saturation parameter Beam A1
    return Ix_minus(x, y, z)/Is
end

function sB2(x, y, z) #saturation parameter Beam B2
    return Iy_plus(x, y, z)/Is
end

function sB1(x, y, z) #saturation parameter Beam B1
    return Iy_minus(x, y, z)/Is
end

function sC1(x, y, z) #saturation parameter Beam C1
    return Iz_minus(x, y, z)/Is
end

function sC2(x, y, z) #saturation parameter Beam C2
    return Iz_plus(x, y, z)/Is
end

#Simulation time :

t =  10*10^(-3)#s #time of the simulation
dt = 0.1/γ; #time step
Nt = convert(Int, round(t/dt)); #number of iteration
```

```julia
#Velocity vectors are taken randomly following a Maxwell-Boltzmann distribution over space
#Equivalent : Each component of each velocity vector is taken randomly following a Gaussian distribution over space

d = Normal() #Normal distribution
k_B = 1.38064852*10^(-23) #m² kg s² K^(-1)
sigma = sqrt(k_B*T/m) #Variance
particles_vx = sigma*rand(d, N);
particles_vy = sigma*rand(d, N);
particles_vz = sigma*rand(d, N);

#Arbitrary Velocity vectors
#particles_vx = -10.0*ones(N)
#particles_vy = zeros(N)
#particles_vz = zeros(N)

#if you want to work with a file
#writedlm(file_speed, particles_vx)
#if filesize(file_speed)==0
#    writedlm(file_speed, [molecules_vx zeros(N,Nt-1)])
#else
#    writedlm(file_speed, [molecules_vx zeros(N,Nt-1); readdlm(file_speed)])
#end

current_speed = [particles_vx, particles_vy, particles_vz]; #speed at t=0
speed_over_time = [] #list storing the speeds
append!(speed_over_time, current_speed)
```

```julia
#In the case of a cavity
#This function defines the behaviour of the atoms in the zone

function bouncing1(system)
    
    position = system[1] #initial positions of the atoms
    speed = system[2] #initial speeds of the atoms
    rectified_position = position[:] #stores the positions after the atoms having rebounced
    rectified_speed = speed[:] #stores the speeds after the atoms having rebouncedd
    
    x = position[1] #initial position along x
    vx = speed[1] #initial velocity along x

    for i = 1:N
    #for each particule

        particule_x = x[i] #extract its position
        particule_vx = vx[i]#extract its speed
        
        if particule_x^2>Bx^2 #The particle is considered outside the cavity if |x| > Bx

            rectified_position[1][i] = sign(particule_x)*(2*Bx-abs(particule_x)) #The extra trip outside the cavity was in fact made inside the cavity 
            rectified_speed[1][i] = -1*particule_vx #the speed flips (the particle rebounds)
        end
        #otherwise the particle is inside the cavity and no changes have to be operated.
    end
    
    return [rectified_position,rectified_speed] #return the new system
end 
```

```julia
function mean_squared_speed(speed)
    
    #returns the instantaneous temperature of the system T= m/(degree_of_freedom*N*k_B)*sum_{i=1,N}(v_i^2) 
    #Here we have 1 degree of freedom
    
    v2 = 0
    
    for i in 1:N
    
        v2 = v2 + speed[1][i]^2 + speed[2][i]^2 + speed[3][i]^2
    
    end
    
    return 1/N*v2
end
```

```julia
squared_speed_over_time = [mean_squared_speed(current_speed)]
```

```julia
function mean(speed)

  #compute the mean of the system

  mean = [0,0,0]

  for i in 1:N

    mean = mean + [speed[1][i],speed[2][i],speed[3][i]]

  end

  return mean/N
  
end 
```

```julia
mean_over_time = []
append!(mean_over_time, [mean(current_speed)])
```

```julia
variance_over_time = [mean_over_time[1][1]^2 + mean_over_time[1][2]^2 + mean_over_time[1][3]^2 - squared_speed_over_time[1]]
```

```julia
time = [0.0] #will store the time

for j = 1 : Nt #at each time step
    
    instant_t = j*dt #s #time since the beginning of the simulation
    append!(time, instant_t)
        
    former_position = current_position[:] #the previous position was obtained at t-dt
    former_speed = current_speed[:] #the previous speed was obtained at t-dt
    
    current_position = [zeros(N), zeros(N), zeros(N)] #the position of each particle at time t is initiated at zero
    current_speed = [zeros(N), zeros(N), zeros(N)] #the speed of each particle at time t is initiated at zero
    
    for i = 1:N 
    #for each particule
        
        #extract the position at time t-dt
        x = former_position[1][i] 
        y = former_position[2][i]
        z = former_position[3][i]

        #extract the speed at time t-dt
        vx = former_speed[1][i]
        vy = former_speed[2][i]
        vz = former_speed[3][i] 

        new_speed_x = vx #the new speed is initiated as the same as the one at t-dt
        new_speed_y = vy
        new_speed_z = vz
        
        ###Beam A1###

        rho_A1 = sA1(x,y,z)/2/(1+sA1(x,y,z)+(2*(δ+k*vx)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam A1 emits in the direction -e_x sigma - polarized light

        γpA1 = γ*rho_A1 #Total scattering rate at time t
        PeeA1 = dt*γpA1 #Probability that a Photon has been scattered during a time dt
        rvaA1 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]  
        
        if rvaA1<PeeA1 #occures with probability Pee1
            
            #random_unitary_vector = [sin(thetaA1)*cos(phiA1), sin(thetaA1)*sin(phiA1), cos(thetaA1)]
            thetaA1 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiA1 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(-1+sin(thetaA1)*cos(phiA1)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*sin(thetaA1)*sin(phiA1)
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*cos(thetaA1)
            
        end

        ###Beam A2###

        rho_A2 = sA2(x,y,z)/2/(1+sA2(x,y,z)+(2*(δ-k*vx)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam A2 emits in the direction e_x sigma + polarized light

        γpA2 = γ*rho_A2 #Total scattering rate at time t
        PeeA2 = dt*γpA2 #Probability that a Photon has been scattered during a time dt
        rvaA2 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]  
        
        if rvaA2<PeeA2 #occures with probability PeeA2
            
            #random_unitary_vector = [sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)]
            thetaA2 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiA2 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(1+sin(thetaA2)*cos(phiA2)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*sin(thetaA2)*sin(phiA2)
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*cos(thetaA2)
        end

        ###Beam B1###

        rho_B1 = sB1(x,y,z)/2/(1+sB1(x,y,z)+(2*(δ+k*vy)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam B1 emits in the direction -e_y sigma - polarized light

        γpB1 = γ*rho_B1 #Total scattering rate at time t
        PeeB1 = dt*γpB1 #Probability that a Photon has been scattered during a time dt
        rvaB1 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]
      
        if rvaB1<PeeB1 #occures with probability Pee1
            
            #random_unitary_vector = [sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)]
            thetaB1 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiB1 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaB1)*cos(phiB1)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*(-1+sin(thetaB1)*sin(phiB1))
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*cos(thetaB1)
            
        end
        
        ###Beam B2###

        rho_B2 = sB2(x,y,z)/2/(1+sB2(x,y,z)+(2*(δ+k*vy)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam B2 emits in the direction +e_y sigma + polarized light

        γpB2 = γ*rho_B2 #Total scattering rate at time t
        PeeB2 = dt*γpB2 #Probability that a Photon has been scattered during a time dt
        rvaB2 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]

        if rvaB2<PeeB2 #occures with probability Pee1
            
            #random_unitary_vector = [sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)]
            thetaB2 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiB2 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaB2)*cos(phiB2)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*(1+sin(thetaB2)*sin(phiB2))
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*cos(thetaB2)
            
        end

        ### Beam C1 ###
        
        rho_C1 = sC1(x,y,z)/2/(1+sC1(x,y,z)+(2*(δ+k*vy)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam C1 emits in the direction -e_z sigma - polarized light

        γpC1 = γ*rho_C1 #Total scattering rate at time t
        PeeC1 = dt*γpC1 #Probability that a Photon has been scattered during a time dt
        rvaC1 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]

        if rvaC1<PeeC1 #occures with probability Pee1
            
            #random_unitary_vector = [sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)]
            thetaC1 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiC1 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaC1)*cos(phiC1)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaC1)*sin(phiC1))
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*(-1+cos(thetaC1))
            
        end

        ### Beam C2 ###

        rho_C2 = sC2(x,y,z)/2/(1+sC2(x,y,z)+(2*(δ+k*vy)/γ)^2) #Probability of finding the two-level atom in an excited state at time t
        #Beam C2 emits in the direction +e_z sigma + polarized light

        γpC2 = γ*rho_C2 #Total scattering rate at time t
        PeeC2 = dt*γpC2 #Probability that a Photon has been scattered during a time dt
        rvaC2 = rand(Float64) #random variable 1, following a uniform distribution on [0,1]

        if rvaC2<PeeC2 #occures with probability Pee1
            
            #random_unitary_vector = [sin(theta)*cos(phi), sin(theta)*sin(phi), cos(theta)]
            thetaC2 = 2*pi*rand(Float64) #random inclination of the emitted photon
            phiC2 = 2*pi*rand(Float64) #random azimuth of the emitted photon
            
            #v[t+dt] = v[t] + hbar*k/m(ex+random_unitary_vector)
            new_speed_x = new_speed_x + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaC2)*cos(phiC2)) 
            new_speed_y = new_speed_y + 6.62607004/2/pi*10^(-34)/m*k*(sin(thetaC2)*sin(phiC2))
            new_speed_z = new_speed_z + 6.62607004/2/pi*10^(-34)/m*k*(1+cos(thetaC2))
            
        end

        #the speed of particle i is changed only if the atom has emitted a photon
        current_speed[1][i] = new_speed_x 
        current_speed[2][i] = new_speed_y
        current_speed[3][i] = new_speed_z
        
        current_position[1][i] = x + new_speed_x*dt #in all cases, the position is changed 
        current_position[2][i] = y + new_speed_y*dt
        current_position[3][i] = z + new_speed_z*dt
    end
        
    #if the experiment is done in a cavity, uncomment the following
    #system_to_bounced = [current_position, current_speed]; #system formed by position and speed at t
    #system_bounced = bouncing1(system_to_bounced); #replace the particles inside the cavity
    #current_position = system_bounced[1]; #get the corrected position
    #current_speed = system_bounced[2]; #get the corrected speed
    
    append!(position_over_time,current_position); 
    append!(speed_over_time, current_speed);
    append!(squared_speed_over_time , mean_squared_speed(current_speed));
    append!(mean_over_time, [mean(current_speed)])

    append!(variance_over_time, mean_over_time[j+1][1]^2 + mean_over_time[j+1][2]^2 + mean_over_time[j+1][3]^2 - squared_speed_over_time[j+1]);
    
    
end
```

```julia
#Method 1 : Sliding Mean
#Step A : smooth the the mean_over_time

p = 1000 #smooth parameter
proper_mean_speed = []

for j in p:Nt+1

   append!(proper_mean_speed, sum(mean_over_time[j-p+1:j])/p)

end
png(plot(time[p:Nt+1],proper_mean_speed, xlabel = "t (s)", ylabel = "<v(t)> (m.s-1)"), "Mean_speed_over_time_filtered")
```

```julia
#Step B : Compute its derivative over time

derivative_mean_speed = []

for j in 2:Nt+2-p

    append!(derivative_mean_speed,(proper_mean_speed[j]-proper_mean_speed[j-1])/dt)

end
```

```julia
#Step C : We are interested in comparing the computed derivative of the speed with the theory
expected_derivative = []
for speed in proper_mean_speed[2:Nt+2-p]
    append!(expected_derivative, 1/m*h/2/pi*k*γ/2*s(0)*(1/(1+s(0)+(2*(δ-k*speed)/γ)^2)))
end
```

```julia
#Step D : Show the unfiltered derivative and compare it with the theory
#Plot the derivative of the speed as a function of the speed
#Plot the expected value of the Fore 
plt1 = plot(proper_mean_speed[2:Nt+2-p], derivative_mean_speed, label = "computed", title = "Derivative of mean speed in function of mean speed", xlabel = "<v>(t) (m.s-1)", ylabel = "d<v(t)>/dt (m/s)")
plot!(proper_mean_speed[2:Nt+2-p], expected_derivative, label = "theory") 
png(plt1, "Forces_1D1Laser")
```
